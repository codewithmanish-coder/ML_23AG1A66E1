```python
import pandas as pd
data=pd.read_csv('Name-Age-Gender-Marks-City.csv')
data.head()
```

```python
data.fillna({
    'AGE':data['AGE'].mean(),
    'MARKS':data["MARKS"].mean(),
    'CITY':'Unknown'
},inplace=True)
print(data)
```

```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Alice'],
    'Age': [18, 18, 18],
    'Gender': ['F', 'M', 'F'],
    'Marks': [85, 90, 85]
}
df = pd.DataFrame(data)
print("Original Data:")
print(df)
df_no_duplicates = df.drop_duplicates()
print("\nAfter Removing Duplicates:")
print(df_no_duplicates)
```
